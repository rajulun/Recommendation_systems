1) Детерминированные алгоритмы item-item - предсказывают те числа, которые стоят в матрице, в item-item алгоритмах используется KNN, У item-item алгоритмов большая сложность, на предсказание у этих алгоритмов уходит много времени, поэтому здесь стараються удалить непопулярные товары в предфильтрации, в разы сократив таким образом количество товаров.    
ALS - лишь способ оптимизации (поиска коэффициентов в эмбеддингах). Здесь реальные значения переводятся в некие эмбеддинги товаров и юзеров. ALS намного быстрее item-item. Используется регуляризация. Еще плюс ALS в том что в начале дня можно зафиксировать за прошедший день эмбеддинги товаров, и в течении дня добавлять эмбеддинги юзеров при совершении сделок юзерами.
