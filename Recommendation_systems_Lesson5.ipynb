1 В чем принципиальное отличие гибридных рекомендательных систем от коллаборативной филтьтрации?

Коллаборативная фильтрация - взаимодействие юзера с различными товарами.   
Гибридные рекомендательные системы - коллаборативная информация + контентная информация (внешние фичи юзеров и внешние фичи товаров)

2 Приведите 2-3 примера задач, в которых необходимо использовать гибридные системы.

а) Холодный старт юзера - когда появился новый юзер, и о нем мало информации. В таком случае мы используем 
внешние фичи этого юзера, такие как возраст, частота появления в магазине, как расплачивался, средний чек и т.д.   
б) Холодный старт товара -когда появился новый товар, и о нем мало информации. В таком случае мы используем 
внешние фичи этого товара, например, вес, срок годности, цену и т.д.
